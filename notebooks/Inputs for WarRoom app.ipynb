```python
import os
import pandas as pd
import numpy as np
import json
import re
from ppa.cascade import Cascade
```

```python
dat = pd.read_csv('data/cascade/d_cascade_2019.csv')
dat = {row['State']: row for _, row in dat.iterrows()}
```

```python
files = os.listdir('out/cf_pars')
n_sel = 50
```

```python
js = dict()

for file in files:
    loc = file.split('pars_')[1].split('.json')[0]

    cas = json.load(open(f'out/cf_pars/{file}'))
    
    cas = cas[:n_sel]
    js[loc] = cas
```

```python
with open('docs/pars_cascade_wr.json', 'w') as f:
    json.dump(js, f)
```

```python
def fmt(cs_json, d):
    cs = Cascade.parse_json(cs_json)
    tar = cs.match_targets()
    dist = np.power(cs.match_targets()['CNR_Pub'] - d.CNR_pub, 2)
    dist += np.power(cs.match_targets()['CNR_Eng'] - d.CNR_eng, 2)

    return {
        'Pars': cs_json,
        'Tar': tar,
        'Dist': dist,
    }
```

```python
file = files[0]
file
loc = file.split('pars_')[1].split('.json')[0]

d = dat[loc]
cas = json.load(open(f'out/cf_pars/{file}'))

cas = [fmt(cs, d) for cs in cas]

cas
```

```python
max_a_post = min(cas, key=lambda x: x['Dist'])
max_a_post['Data'] = d.to_dict()

```
